## Task-1: GenAI- Prompt Templates Using Lang Chain

## Objective

The objective of this assignment is to build a dynamic and reusable prompt system using LangChain PromptTemplate instead of hardcoded prompts. The goal is to design structured prompt pipelines that accept user inputs and generate flexible prompts for different use cases. This assignment focuses on implementing multi-input templates, chat-based prompting, input validation, and reusable prompt generation for real-world Generative AI applications.

## Step-1: Install and Import Libraries

```python
!pip install langchain==0.0.350

from langchain.prompts import PromptTemplate
from langchain.prompts.chat import ChatPromptTemplate
```

# Step-1: Replace Hardcoded Prompts

```python
template = "Explain {topic} in simple terms for beginners"

prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

output = prompt.format(topic="Artificial Intelligence")
print(output)
```

# Step-2: Multi-Input Prompt System 


```python
template = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test cases
print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

# Task 3: Prompt Variations Engine

```python

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

# Task 4: ChatPromptTemplate System

```python


chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

print(messages)
```

# Task 5: Input validation Layer

```python


def validate_inputs(audience, tone):
    
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]
    
    if audience not in valid_audience:
        audience = "beginner"
        
    if tone not in valid_tone:
        tone = "formal"
        
    return audience, tone


# Test
print(validate_inputs("beginner", "fun"))
print(validate_inputs("kids", "friendly"))
```

# Task 6: Prompt Generator App

```python


def generate_prompt(topic, audience, tone, style):
    
    audience, tone = validate_inputs(audience, tone)
    
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions on {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }
    
    template = templates.get(style)
    
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )
    
    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )


print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

# Task 7:  Template Reusability test

```python


template = "Explain {topic} for {audience} in a {tone} tone"

reusable_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

test_inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "kids", "casual"),
    ("Data Science", "students", "formal"),
    ("Deep Learning", "engineers", "technical"),
    ("NLP", "researchers", "formal")
]

for topic, audience, tone in test_inputs:
    print(reusable_prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```
